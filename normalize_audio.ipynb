```python
import os
import subprocess
import pandas as pd
from pydub.utils import mediainfo

gen_folder_path_old = '/Users/jacksloan/ML_Projects/urban_sound_classification/us8k/UrbanSound8k/UrbanSound8k/audio/fold'
gen_folder_path_1 = "/Users/jacksloan/ML_Projects/urban_sound_classification/us8k/UrbanSound8k/UrbanSound8k/audio_step1/fold"
gen_folder_path_2 = '/Users/jacksloan/ML_Projects/urban_sound_classification/us8k/UrbanSound8k/UrbanSound8k/audio_step2/fold'
gen_folder_path_new = '/Users/jacksloan/ML_Projects/urban_sound_classification/us8k/UrbanSound8k/UrbanSound8k/audio_cleaned/fold'
```

```python
import math
def get_multiplier(old_db, new_db):
    delta_db = new_db - old_db
    return 10 ** (delta_db / 20)
```

```python
df = pd.read_excel("/Users/jacksloan/ML_Projects/urban_sound_classification/Audio_Metadata_0.xlsx")

df.head()
```

```python
avg = round(df.iloc[0,4])

folder_path_old = "x"
folder_path_1 = "x"
folder_path_2 = "x"
folder_path_new = "x"

i = 0
for index, row in df.iterrows():
    #print(f'{row["Folder"][-1]}')
    if row["Folder"][-1] != folder_path_old[-1]:
        print(f'{row["Folder"][-1]} AAAAA {folder_path_old[-1]}')
        i += 1
        folder_path_old = f"{gen_folder_path_old}{i}"
        folder_path_1 = f"{gen_folder_path_1}{i}"
        folder_path_2 = f"{gen_folder_path_2}{i}"
        folder_path_new = f"{gen_folder_path_new}{i}"
        
    sr_command = ['ffmpeg', '-i', f'{folder_path_old}/{row["Filename"]}', '-ar', '44100', f'{folder_path_1}/{row["Filename"]}']
    
    #Sox was distorting 24 bit files
    bd_command = ['ffmpeg', '-i', f'{folder_path_1}/{row["Filename"]}', '-c:a', 'pcm_s16le', f'{folder_path_2}/{row["Filename"]}']

    #calculated in the excel file 0
    avg_volume = -25.1
    file_volume = float(row["Mean Volume"])

    multiplier = get_multiplier(file_volume, avg_volume)
    
    vol_command = ['ffmpeg', '-i', f'{folder_path_2}/{row["Filename"]}', '-filter:a', f'volume={multiplier}', f'{folder_path_new}/{row["Filename"]}']
    
    subprocess.run(sr_command, capture_output=True, text=True)
    subprocess.run(bd_command, capture_output=True, text=True)
    subprocess.run(vol_command, capture_output=True, text=True)

print("Done")
    
    
        
```
